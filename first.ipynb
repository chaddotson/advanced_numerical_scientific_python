# plotting the trig functions


```python
import math

x = [((i * 2.0 * math.pi)/100.0) - math.pi for i in range(101)]
print(x)
```

```python
import math

import numpy as np

x = np.linspace(-math.pi, math.pi, 100)
```

```python
y1 = [math.cos(i) for i in x]
y2 = [math.sin(i) for i in x]
y3 = [math.tan(i) for i in x]
```

```python
plot1 = mpl.plot(x, y1, linewidth=3, color='r')
plot2 = mpl.plot(x, y2, linewidth=3, color=(0,0,0))
#plot3 = mpl.plot(x, y3)
mpl.title("trig functions")
mpl.show()
```

# rectangular wave function

```python
x = [i / (100.0 - 1.0) for i in range(100)]
y = [math.sin((2.0 * math.pi * xx) / (0.5)) for xx in x]
rect = [+1 if yy > 0 else -1 for yy in y]
print(rect)
```

```python
plot1 = mpl.plot(x, rect, linewidth=3, color='r')
mpl.show()
```

fourier expansion

$$f(x) = \frac(4)(\pi)\sum_(n=1,3,5,....


```python
y= [ 0.0 for i in range(100)]
for n in range(1, 8, 2):
    for i in range(100):
        y[i] = y[i] + (4.0 / (math.pi * n)) * math.sin((2.0 * math.pi * n * x[i]) / (0.5))
    mpl.plot(x, y)
mpl.title("First few terms of rectangular wave")
mpl.xlabel("$x$")
mpl.ylabel("$y$")
mpl.show()

```
