```python
import matplotlib.pyplot as mpl
import numpy as np
from scipy.optimize import minimize
```

```python
num_levels = 35
# z = 0.5 * (1-X)**2 + (Y-X**2)**2
def f(x): # objective function... quadratic surface
    return 0.5 * (1-x[0])**2 + (x[1]-x[0]**2)**2

X, Y = np.meshgrid(np.linspace(0, 2, 100), np.linspace(0, 2, 100))
cp = mpl.contourf(X, Y, f([X, Y]), num_levels, cmap=mpl.cm.jet) # also valid
cp2 = mpl.contour(X, Y, f([X, Y]), num_levels, cmap=mpl.cm.gray)
#z = 0.5 * (1-X)**2 + (Y-X**2)**2
#cp = mpl.contourf(X, Y, z, num_levels, cmap=mpl.cm.jet)
#cp2 = mpl.contour(X, Y, z, num_levels, cmap=mpl.cm.gray)
mpl.colorbar(cp)
mpl.show()
```
