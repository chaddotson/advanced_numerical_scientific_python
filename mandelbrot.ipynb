```python
import matplotlib.pyplot as mpl
import numpy as np

image_number = np.complex(0, 1) # python also supports complex out of the box
def mandelbrot(h, w, iterations=20):
    y = np.linspace(-1.4, 1.4, h)
    x = np.linspace(-2.0, 0.8, w)
    Y, X = np.meshgrid(y, x)
    c = X + Y * image_number
    z = c
    divtime = np.zeros(z.shape, dtype=int) + iterations
    for i in range(iterations):
        z = z**2 + c
        diverge = z * np.conj(z) > 2**2
        div_now = diverge & (divtime==iterations)
        divtime[div_now] = i
        z[diverge] = 2
        
    return divtime
```

```python
mpl.imshow(mandelbrot(400, 400, 20), cmap=mpl.cm.jet)
mpl.show()
```
