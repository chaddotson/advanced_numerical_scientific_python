```python
x = np.linspace(0, 1, 100)
y = np.random.randint(-16, 16, 100)

plot1 = mpl.scatter(x, y, marker="^", color='r' )
mpl.show()
plot2 = mpl.plot(x, y, marker="^", markersize=12, markerfacecolor=(1,1,1), markeredgecolor=(0,0,0), color='r')
mpl.show()
```

```python
fig = mpl.figure(figsize=(7,7))
x = np.linspace(0, 2 * np.pi, 100)
axes = fig.add_subplot(3,1,1)
axes.plot(x, np.sin(x))

axes = fig.add_subplot(3,1,2)
axes.plot(x, np.cos(x))

axes = fig.add_subplot(3,1,3)
axes.plot(x, x**2)

mpl.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = mpl.figure(figsize=(12, 7))
ax = fig.add_subplot(1, 1, 1, projection="3d")
xs = np.linspace(0.0, 50.0, 20)
ys = np.linspace(0.0, 20.0, 20)
z = np.random.randint(-12, 12, size=(20))
z2 = np.random.randint(-16, 16, size=(20))
ax.scatter(xs, ys, z, marker="^", color="r")
ax.scatter(xs, ys, z2, marker="o", color="r")
mpl.show()
```

```python
x, y = np.linspace(-1.0, 2.5, 100), np.linspace(-1.5, 1.5, 100)
X, Y = np.meshgrid(x, y)
z = (-5*X**3+2*Y**2+4.0*X)

from mpl_toolkits.mplot3d import Axes3D
fig = mpl.figure(figsize=(12, 7))
ax = fig.add_subplot(1, 1, 1, projection="3d")

p1 = ax.contourf(X, Y, z, 25)
p2 = ax.contour(X, Y, z, 25, cmap=mpl.cm.gray)

mpl.colorbar(p1)

mpl.show()
```

```python
def saveSnapshot(matr, save=False, fn="default"):
    img = mpl.imshow(matr, interpolation="nearest", cmap=mpl.cm.Blues_r)
    
    if save:
        mpl.savefig(fn)
    else:
        mpl.show()
        
matr = np.random.random((64, 64))

saveSnapshot(matr, save=True, fn="tester.jpg")
        
```
