```python
import math
x = [i / (100.0 - 1.0) for i in range(100)]
y = [ 0.0 for i in range(100)]
for n in range(1, 8, 2):
    for i in range(100):
        y[i] = y[i] + (4.0 / (math.pi * n)) * math.sin((2.0 * math.pi * n * x[i]) / (0.5))
    mpl.plot(x, y)
mpl.title("First few terms of rectangular wave")
mpl.xlabel("$x$")
mpl.ylabel("$y$")
mpl.show()
```

```python
x = np.linspace(0.0, 1.0, 100)
y = np.zeros(100)
for n in range(1, 8 ,2):
    y+= (4.0 / (np.pi * n)) * np.sin((2 * np.pi * n * x) / 0.5)
    mpl.plot(x,y)
mpl.show()
```
