```python
import matplotlib.pyplot as mpl
import numpy as np
```

```python
THRESHOLD = 10


def julia(h, w, c, iterations=255):
    x, y = np.linspace(-2, 2, w), np.linspace(-2, 2, h)
    
    z = np.zeros((w, h))
        
    for i, re in enumerate(x):
        for j, im in enumerate(y):
            zz = complex(re, im)
            n = iterations
            while abs(zz) < THRESHOLD and n >= 0:
                zz = zz * zz + c
                n-=1
                
            z[i, j] = n
    return z
```

```python
fig = mpl.figure(figsize=(15,15))
ax = fig.add_subplot(1, 1, 1)
g = ax.imshow(julia(600, 600, complex(0, 0.75), 255)[::-1], cmap=mpl.cm.gray, origin="lower")

mpl.colorbar(g)
mpl.show()
```
