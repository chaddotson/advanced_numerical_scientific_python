```python
import matplotlib.pyplot as mpl
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import odeint
```

```python
def lorenz(x, y, z, r=28, s=10, b=2.667):
    xdot = s*(y-x)
    ydot = r*x-y-x*z
    zdot = x*y-b*z
    return xdot, ydot, zdot
    
```

```python
dt = 0.01
tmax = 10000

xs, ys, zs = np.zeros((tmax+1)), np.zeros((tmax+1)), np.zeros((tmax+1))

xs[0], ys[0], zs[0] = 0.0, 1.0, 1.05
for t in range(tmax):
    xdot, ydot, zdot = lorenz(xs[t], ys[t], zs[t])
    
    xs[t+1] = xs[t]+(xdot * dt)
    ys[t+1] = ys[t]+(ydot * dt)
    zs[t+1] = zs[t]+(zdot * dt)
    
print("Done")





```

```python
fig = mpl.figure(figsize=(13,13))
axes = fig.add_subplot(1,1,1, projection='3d')
axes.plot(xs, ys, zs, lw=0.5)
mpl.show()
```

```python
def lorenz2(xyz, t0, s=10, b=2.667, r=28):
    x,y,z = xyz
    return [s*(y-x), x*(r-z) - y, x*y-b*z]


x0 = [0, 1, 1.05]

t = np.linspace(0, 100, 10000)
x_t = odeint(lorenz2, x0, t)
print(x_t)
```

```python
fig = mpl.figure(figsize=(13,13))
axes = fig.add_subplot(1,1,1, projection='3d')
axes.plot(x_t[:,0], x_t[:,1], x_t[:,2], lw=0.5)
mpl.show()
```
