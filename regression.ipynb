```python
import matplotlib.pyplot as mpl
import numpy as np
import scipy
```

```python
x = np.linspace(0.0, 1.0, 100)
yline = 0.5 *x

y = yline + (np.random.random(x.size) * 0.20 - 0.10)
mpl.plot(x, yline)
mpl.plot(x, y, marker='^', markerfacecolor='w', markeredgecolor='k', color='r')
mpl.show()
```

```python
fit = np.polyfit(x, y, 1)
print(fit)
```

```python
mpl.plot(x, np.polyval(fit, x))
mpl.plot(x, yline, linestyle='--')
mpl.scatter(x, y, marker='^')
mpl.show()
```
